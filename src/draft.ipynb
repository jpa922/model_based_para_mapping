```python
%matplotlib inline
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np

X_train = np.arange(0,100,0.1)
y_train = 0.01*(X_train)**2

X_test = np.arange(0.05,100,0.1)
y_test = 0.01*(X_test)**2

X_train=X_train.reshape(X_train.shape[0],1)        #had tough time without this,
y_train=y_train.reshape(y_train.shape[0],1)        #had tough time without this,
X_test=X_train.reshape(X_test.shape[0],1)        #had tough time without this,
y_test=y_test.reshape(y_test.shape[0],1)        #had tough time without this,
```

```python
model = Sequential()
model.add(Dense(1, 100, init='lecun_uniform'))
model.add(Activation('tanh'))
# model.add(Dropout(0.5))
model.add(Dense(100, 100, init='lecun_uniform'))
model.add(Activation('tanh'))
# model.add(Dropout(0.5))
model.add(Dense(100, 100, init='lecun_uniform'))
model.add(Activation('tanh'))
# model.add(Dropout(0.5))
model.add(Dense(100, 1, init='lecun_uniform'))
model.add(Activation('linear'))

sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='mean_squared_error', optimizer=sgd)

model.fit(X_train, y_train, nb_epoch=500, batch_size=10)
# score = model.evaluate(X_test, y_test, batch_size=10)
```

```python
y_predict = model.predict(X_test,batch_size=10,verbose=1)
plt.plot(X_test,y_test, 'r')
plt.plot(X_test,y_predict, 'b.')
# plt.axis([0, 6, 0, 20])
plt.show()
```
